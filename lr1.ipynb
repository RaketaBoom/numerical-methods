### 5 Решить систему уравнений

```python
#numpy
M2 = np.array([[3, 2, 1], [3, 3, 2], [5, 5, 3]], int) # Матрица (левая часть системы)
v2 = np.array([5, 7, 11], int) # Вектор (правая часть системы)


print(np.linalg.solve(M2, v2))
```

```python
#sympy
M2_s = sympy.Matrix(M2)
V2_s = sympy.Matrix(v2)

result = M2_s.solve(V2_s)

print(result)
```

### 6 Вычислить интеграл

$$ \int ^{\pi/4} _0 \frac 1 {1 + 2\sin^2x}  dx$$

```python
# #sympy
# Определяем символьную переменную
x = sympy.symbols('x')

# Определяем функцию
f = 1 / (1 + 2*sympy.sin(x)**2)

# Вычисляем интеграл
integral_sym = sympy.integrate(f, (x, 0, sympy.pi/4))
# print("Интеграл с помощью SymPy:", integral_sym)
# Оценка численного значения интеграла
numerical_value = integral_sym.evalf()
print("Численное значение интеграла:", numerical_value)

```

```python
#scipy
# Определяем функцию
def f(x):
    return 1 / (1 + 2*np.sin(x)**2)

# Вычисляем интеграл
integral_scipy, _ = integrate.quad(f, 0, np.pi/4)
print("Интеграл с помощью SciPy:", integral_scipy)
```

### 7 Задание

$$\int_{-1}^2{dy}\int_{y^2}^{y+2} {y^2}dx$$

```python
#sympy
import sympy as sp

# Определяем символьные переменные
x, y = sp.symbols('x y')

# Определяем функцию
f = y**2

# Вычисляем интеграл
integral_sym = sp.integrate(f, (x, y**2, y+2), (y, -1, 2))
print("Интеграл с помощью SymPy:", integral_sym)
```

```python
#scipy
from scipy.integrate import dblquad

# Определяем функцию
def f(x, y):
    return y**2

# Вычисляем интеграл
integral_scipy, _ = dblquad(f, -1, 2, lambda y: y**2, lambda y: y+2)
print("Интеграл с помощью SciPy:", integral_scipy)
```
